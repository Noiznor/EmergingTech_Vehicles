# Complex Convolutional Neural Network (CNN)
  Three Convolutional Blocks: It has three distinct Conv2D and MaxPooling2D pairs. This deeper structure allows it to learn more intricate patterns from the images. The first block might learn simple edges, the second might learn shapes, and the third could learn more complex object parts.

  Dropout Layer: It includes a Dropout(0.5) layer. This is a regularization technique that helps prevent the model from "memorizing" the training data (overfitting), which should help it perform better on new, unseen images.

  -------------------------------------------------

```python
import tensorflow as tf
import tensorflow_datasets as tfds
from tensorflow.keras import layers, models
import numpy as np

# ===================================================================
# STEP 1: LOAD AND FILTER THE DATASET
# ===================================================================
(ds_train, ds_test), ds_info = tfds.load(
    'cifar100', split=['train', 'test'], shuffle_files=True, as_supervised=True, with_info=True)

VEHICLE_1_LABELS = [8, 13, 48, 58, 90]
def is_vehicle(image, label):
  return tf.reduce_any(tf.equal(label, VEHICLE_1_LABELS))

ds_train_vehicles = ds_train.filter(is_vehicle)
ds_test_vehicles = ds_test.filter(is_vehicle)
print("✅ Dataset loaded and filtered.")

# ===================================================================
# STEP 2: CREATE THE CORRECTED DATA PIPELINE
# ===================================================================
VEHICLE_1_LABELS_TENSOR = tf.constant(VEHICLE_1_LABELS, dtype=tf.int64)
MAPPED_LABELS_TENSOR = tf.constant([0, 1, 2, 3, 4], dtype=tf.int64)
BATCH_SIZE = 32

label_lookup_table = tf.lookup.StaticHashTable(
    initializer=tf.lookup.KeyValueTensorInitializer(keys=VEHICLE_1_LABELS_TENSOR, values=MAPPED_LABELS_TENSOR),
    default_value=-1)

def preprocess(image, label):
  image = tf.cast(image, tf.float32) / 255.0
  label = label_lookup_table.lookup(label)
  return image, label

train_data = ds_train_vehicles.map(preprocess).shuffle(1000).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
test_data = ds_test_vehicles.map(preprocess).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
print("✅ Data pipeline is ready!")

# ===================================================================
# STEP 3: BUILD AND COMPILE THE DEEPER CNN MODEL (UPDATED)
# ===================================================================
model = models.Sequential([
    layers.Input(shape=(32, 32, 3)),

    # First convolutional block
    layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
    layers.MaxPooling2D((2, 2)),

    # Second convolutional block
    layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
    layers.MaxPooling2D((2, 2)),

    # Third convolutional block (NEW)
    layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
    layers.MaxPooling2D((2, 2)),

    # Flatten the output
    layers.Flatten(),

    # Dense layer and Dropout layer (NEW)
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5), # Drops 50% of neurons to prevent overfitting

    # Output layer
    layers.Dense(5, activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()

# ===================================================================
# STEP 4: TRAIN THE MODEL FOR MORE EPOCHS (UPDATED)
# ===================================================================
print("\nStarting model training...")
history = model.fit(
    train_data,
    epochs=30, # Increased from 15 to 30
    validation_data=test_data
)
print("\n✅ Training complete!")

# ===================================================================
# STEP 5: EVALUATE THE FINAL MODEL ACCURACY
# ===================================================================
print("\nEvaluating the final model on the test data...")
final_loss, final_accuracy = model.evaluate(test_data)
print(f"\nFinal Test Loss: {final_loss:.4f}")
print(f"Total Accuracy on Test Set: {final_accuracy*100:.2f}%")

import matplotlib.pyplot as plt

# ===================================================================
# STEP 6: PLOT TRAINING AND VALIDATION ACCURACY AND LOSS
# ===================================================================
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs_range = range(len(acc))

plt.figure(figsize=(12, 4))

# Accuracy plot
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

# Loss plot
plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')

plt.show()
```
